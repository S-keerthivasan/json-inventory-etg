```python
import json
from IPython.display import clear_output
from time import ctime



def addnewitem(datasheet):
    tot=int(input("Enter no of items to be added"))
    for i in range(tot):
        uid=input("Enter product uid")
        tempdict={"name":"no value","instock":0,"price":0}
        name1=input("product name ?")
        price=int(input("product price ?"))
        tempdict["name"]=name1
        tempdict["price"]=price 
        datasheet[uid]=tempdict
    
def addstock(datasheet):
    tot=int(input("Enter no of items to be added"))
    for i in range(tot):
        uid=input("Enter product uid")
        if(uid in datasheet.keys()):
            n=int(input("Quantity to be added"))
            datasheet[uid]["instock"]+=n
        else:
            print("Item does not exist, add item first")
        
def billitems(datasheet):
    with open("history.json","r") as tempin:
        dict1=json.load(tempin)
    l=[]    
    tot=int(input("No of items to be billed ?"))
    pricesum=0
    i=1
    while(i<=tot):
        uid=input("Enter uid of product")
        quan=int(input("Enter quantity"))
        if(datasheet[uid]["instock"]>=quan):
            print(datasheet[uid]["name"],"   :   ",datasheet[uid]["price"])
            pricesum+=(datasheet[uid]["price"]*quan)
            datasheet[uid]["instock"]-=quan
            datasheet[uid]["updated time"]=ctime()
            l.append(datasheet[uid]["name"])
            l.append(datasheet[uid]["price"])
            l.append(quan)
        else:
            print("out of stock . only ",datasheet[uid]["instock"],"left in stock")
            i-=1
        i+=1    
    print("value of bag",pricesum)
    l.append(pricesum)
    dict1[ctime()]=l
    with open("history.json","w") as tempout:
        json.dump(dict1,tempout)
            
    
    
def viewstock(datasheet):
    for i in datasheet.items():
        print(i[0],i[1])

datasheet={}

user=input("Enter y to access already existing file \n Enter n to create a new file")
if(user.lower()=="y"):
    filename=input("Enter filename")+".json"
    with open(filename,"r")as infile:
        datasheet=json.load(infile)
else:
    filename=input("Enter filename to create")+".json"
    with open(filename,"w") as outfile:
        json.dump(datasheet,outfile)
    with open(filename,"r")as infile:
        datasheet=json.load(infile)
    dict2={}
    with open("history.json","w") as tempout:
        json.dump(dict2,tempout)
        
flag="y"
while(flag.lower()=="y"):
    choice=int(input("Enter 1 to add new item \n Enter 2 to add stock \n Enter 3 to bill items \n Enter 4 to view stock \n Enter 5 to view purchase history"))
    clear_output(wait=False)
    if(choice==1):
        addnewitem(datasheet)
    elif(choice==2):
        addstock(datasheet)
    elif(choice==3):
        billitems(datasheet)
    elif(choice==4):
        viewstock(datasheet)
    elif(choice==5):
        with open("history.json","r") as tempin:
            dict1=json.load(tempin)
            for i in dict1.items():
                print(i[0],i[1])
    else:
        print("invalid choice") 
    flag=input("Enter y to continue anything else to exit")   
with open(filename,"w")as outfile:
    temp=json.dump(datasheet,outfile)
    
    
```
